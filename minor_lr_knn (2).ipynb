```python
df_train1= pd.read_csv(r"C:\Users\admin\Documents\minor_project\kdd_train.csv")
df_train2= df_test1.drop('service', axis=1)
df_train= pd.get_dummies(df_train2, columns=['protocol_type'])
df_train.head()
```

```python
dupli_train= df_train.copy()
```

```python
dupli_train.replace(change, inplace=True)
dupli_train.head()
```

```python
x_dupli_train= dupli_train.drop('labels', axis=1)
y_dupli_train= dupli_train['labels']
```

```python
x_dupli_test= dupli_test.drop('labels', axis=1)
true_p= dupli_test['labels']
```

```python
#logistic rege
```

```python
x_algo_names = []
y_accuracy = []
y_time = []
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import time
from sklearn.metrics import confusion_matrix
```

```python
start = time.time()
logreg =LogisticRegression()
logreg.fit(x_dupli_train, y_dupli_train)
pred= logreg.predict(x_dupli_test)
cf_matrix_lr = confusion_matrix(true_p, pred)
print("Accuracy: ", accuracy_score(pred, true_p))
elapsed_time_fl = (time.time() - start)

x_algo_names.append("Logistic_Regression")
y_accuracy.append(accuracy_score(pred,true_p))
y_time.append(elapsed_time_fl)
```

```python
#knn classifier
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
start = time.time()
neigh = KNeighborsClassifier()
neigh.fit(x_dupli_train, y_dupli_train)
knnpred = neigh.predict(x_dupli_test)
cf_matrix_knn = confusion_matrix(true_p, knnpred)
print("Accuracy: ", accuracy_score(knnpred, true_p))
elapsed_time_fl = (time.time() - start)

x_algo_names.append("KNN")
y_accuracy.append(accuracy_score(knnpred, true_p))
y_time.append(elapsed_time_fl)
```

```python
y_time
```

```python
from sklearn import tree
```

```python
start = time.time()
DTree = tree.DecisionTreeClassifier()
DTree.fit(x_dupli_train, y_dupli_train)
DTree_ = DTree.predict(x_dupli_test)
cf_matrix_dt = confusion_matrix(true_p, DTree_)
print("Accuracy: ", accuracy_score(DTree_, true_p))
elapsed_time_fl = (time.time() - start)

x_algo_names.append("Decision Tree")
y_accuracy.append(accuracy_score(DTree_, true_p))
y_time.append(elapsed_time_fl)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
start = time.time()
RFclf = RandomForestClassifier()
RFclf.fit(x_dupli_train, y_dupli_train)
RFclf_ = RFclf.predict(x_dupli_test)
cf_matrix_rf = confusion_matrix(true_p, RFclf_)
print("Accuracy: ", accuracy_score(RFclf_, true_p))
elapsed_time_fl = (time.time() - start)

x_algo_names.append("Random Forest")
y_accuracy.append(accuracy_score(RFclf_, true_p))
y_time.append(elapsed_time_fl)
```
